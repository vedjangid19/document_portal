```python
import sys
import traceback
from logger.custom_logger import CustomLogger

logger = CustomLogger().get_logger("app_exception_logger")

class DocumentPortalException(Exception):
    def __init__(self, error_message, error_detail: sys):
        print(error_detail.exc_info())
        _, _, exc_tb = error_detail.exc_info()
        self.filename = exc_tb.tb_frame.f_code.co_filename
        self.line_number = exc_tb.tb_lineno
        self.error_message = str(error_message)
        self.traceback_str = ''.join(traceback.format_exception(*error_detail.exc_info()))

    def __str__(self):
        return f"""
        Error in file: [{self.filename}] at line: [{self.line_number}]
        Error message: [{self.error_message}]
        Traceback: [{self.traceback_str}]
        """

if __name__ == "__main__":
    try:
        a = 1 / 0
    except Exception as e:
        app_exc = DocumentPortalException(e, sys)
        logger.error(app_exc)
        raise app_exc
```
